# Dimension Reduction

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, TSNE, Isomap
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

#    
X, colors = make_swiss_roll(n_samples=800)

# 3D 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=colors, cmap=plt.cm.Spectral)
ax.set_title('3D Swiss Roll')
plt.show()
```

```python

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# MDS
mds = MDS(n_components=2, random_state=0)
X_mds = mds.fit_transform(X)

# Isomap
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X)

# t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)
```

```python

# 2D 
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# PCA
axs[0, 0].scatter(X_pca[:, 0], X_pca[:, 1], c=colors, cmap=plt.cm.Spectral)
axs[0, 0].set_title('PCA')

# MDS
axs[0, 1].scatter(X_mds[:, 0], X_mds[:, 1], c=colors, cmap=plt.cm.Spectral)
axs[0, 1].set_title('MDS')

# Isomap
axs[1, 0].scatter(X_isomap[:, 0], X_isomap[:, 1], c=colors, cmap=plt.cm.Spectral)
axs[1, 0].set_title('Isomap')

# t-SNE
axs[1, 1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=colors, cmap=plt.cm.Spectral)
axs[1, 1].set_title('t-SNE')

plt.tight_layout()
plt.show()
```

```python

#  
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python

# PCA
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

# MDS
mds = MDS(n_components=2, random_state=0)
X_r_mds = mds.fit_transform(X)

# Isomap
isomap = Isomap(n_components=2)
X_r_isomap = isomap.fit_transform(X)

# t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_r_tsne = tsne.fit_transform(X)
```

```python

#  
plt.figure(figsize=(20, 10))

# PCA
plt.subplot(2, 2, 1)
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=2,
                label=target_name)
plt.title('PCA of IRIS dataset')

# MDS
plt.subplot(2, 2, 2)
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], target_names):
    plt.scatter(X_r_mds[y == i, 0], X_r_mds[y == i, 1], color=color, alpha=.8, lw=2,
                label=target_name)
plt.title('MDS of IRIS dataset')

# Isomap
plt.subplot(2, 2, 3)
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], target_names):
    plt.scatter(X_r_isomap[y == i, 0], X_r_isomap[y == i, 1], color=color, alpha=.8, lw=2,
                label=target_name)
plt.title('Isomap of IRIS dataset')

# t-SNE
plt.subplot(2, 2, 4)
for color, i, target_name in zip(['navy', 'turquoise', 'darkorange'], [0, 1, 2], target_names):
    plt.scatter(X_r_tsne[y == i, 0], X_r_tsne[y == i, 1], color=color, alpha=.8, lw=2,
                label=target_name)
plt.title('t-SNE of IRIS dataset')

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

# Tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python

# Iris  .
iris = datasets.load_iris()
X, y = iris.data, iris.target

#      .
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python

#     
criterion='gini'

tree = DecisionTreeClassifier(criterion='gini', random_state=0)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Criterion {criterion}: Test accuracy = {accuracy}')

#    .
plt.figure(figsize=(20, 10))  #   .
plot_tree(tree, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names,
          filled=True)

plt.show()
```

```python

#     

tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
criterion  ='entropy'
print(f'Criterion {criterion}: Test accuracy = {accuracy}')

#    .
plt.figure(figsize=(20, 10))  #   .
plot_tree(tree, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names,
          filled=True)

plt.show()
```

```python

#        .
tree_no_pruning = DecisionTreeClassifier(criterion='gini', random_state=0)
tree_pruning = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=0)
tree_no_pruning.fit(X_train, y_train)
tree_pruning.fit(X_train, y_train)


#   .
y_pred_no_pruning = tree_no_pruning.predict(X_test)
y_pred_pruning = tree_pruning.predict(X_test)
accuracy_no_pruning = accuracy_score(y_test, y_pred_no_pruning)
accuracy_pruning = accuracy_score(y_test, y_pred_pruning)

print(f'No pruning: Test accuracy = {accuracy_no_pruning}')
print(f'With pruning: Test accuracy = {accuracy_pruning}')
```

```python
#    .
plt.figure(figsize=(20, 10))  #   .
plot_tree(tree_no_pruning, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names,
          filled=True)

plt.show()
```

```python
#    .
plt.figure(figsize=(20, 10))  #   .
plot_tree(tree_pruning, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names,
          filled=True)

plt.show()
```
